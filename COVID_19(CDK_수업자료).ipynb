```
# -*- coding: utf-8 -*-
"""Copy of covid-19 탐구 분석 내용 """
 
import pandas as pd  
import plotly.graph_objs as go
import numpy as np
import plotly.express as px
```

```
 
# 전세계 지역별 확진자, 회복자, 사망자 Data Set
url = 'https://raw.githubusercontent.com/datasets/covid-19/master/data/time-series-19-covid-combined.csv'
data = pd.read_csv(url, error_bad_lines=False)
 
# Data Set의 구조 파악
data
grouped = data.groupby('Country/Region')
#df_map = data.groupby('Country/Region').sum().reset_index(drop=False).sort_values('Country/Region')
 
df_map = pd.DataFrame(grouped.sum()['Confirmed']) # DataFrame 
df_map = df_map.reset_index(drop=False).rename(columns= {0:'Country', 1:'Confirmed'})
 
df_map
```

```
data
```

```
# 세계확진자 수 
df = data.groupby('Date').sum().reset_index(drop=False).sort_values('Date')
fig = go.Figure()
fig.add_trace(go.Scatter(
                x=df['Date'],
                y=df['Confirmed'],
                name="Confirmed",
                line_color='deepskyblue',
              ))
fig.add_trace(go.Scatter(
                x=df['Date'],
                y=df['Recovered'],
                name="Recovered",
                line_color='red',
                opacity=1))
fig.add_trace(go.Scatter(
                x=df['Date'],
                y=df['Deaths'],
                name="Deaths",
                line_color='yellow',
                opacity=1))
 
# Use date string to set xaxis range
fig.update_layout(title_text="COVID-19 spreading worldwide")
fig.show()
```

```
# 이탈리아 확진자수 df_Italy 임
df_Italy = data[data['Country/Region'] == 'Italy']
df_Italy = np.column_stack([df_Italy.Date ,df_Italy.Confirmed, df_Italy.Recovered, df_Italy.Deaths])
df_Italy =  pd.DataFrame(df_Italy)
 
df_Italy = df_Italy.reset_index().rename(columns={0: 'date',1: 'confirmed', 2: ' recovered' , 3 : 'deaths'})
df_Italy = df_Italy.drop("index", axis=1)
df_Italy
```
